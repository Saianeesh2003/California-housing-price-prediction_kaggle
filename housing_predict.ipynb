```python
import pandas as pd
data=pd.read_csv('housing.csv')


data
```

```python
data.isnull().sum()
```

```python
median_bedrooms = data['total_bedrooms'].median()
data['total_bedrooms'].fillna(median_bedrooms, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data_encoded = pd.get_dummies(data, columns=['ocean_proximity'], drop_first=True, dtype=int)
```

```python
X = data_encoded.drop('median_house_value', axis=1)
y = data_encoded['median_house_value']
X,y
```

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
y_train_pred=model.predict(X_train)

test_results=pd.DataFrame(
    { 'predicted' : y_pred,
     ' actual ' : y_test }
)
test_results
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
import numpy as np
rmse = np.sqrt(mse) 
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R²) Score: {r2:.4f}")
print(f"Root Mean Squared Error (RMSE): ${rmse:,.2f}")
```
